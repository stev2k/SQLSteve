```r
#
df1 <- read.csv(file="meteorological-data-create.csv", header=TRUE, sep=",")
df1
```

```r
head(sort(df1$Temperature,decreasing=TRUE), n = 10)
```

```r
head(sort(df1$Temperature,decreasing=FALSE), n = 10)
```

```r
library(zoo)
df1$month <- as.Date(as.yearmon(df1$Date.Time))

df1
```

```r
SummarisedTemps <- data.frame(
aggregate(Temperature ~ df1$month, data=df1, FUN=min),
aggregate(Temperature ~ df1$month, data=df1, FUN=max))

SummarisedTemps <- SummarisedTemps[,c(1,2,4)]

colnames(SummarisedTemps) <- c("month", "min", "max")

SummarisedTemps
```

```r
library("ggplot2")
ggplot(SummarisedTemps, aes(month, min)) + geom_line()
   xlab("") + ylab("Temperature")
```

```r
options(repr.plot.width=16, repr.plot.height=9)

plot(SummarisedTemps$month,SummarisedTemps$min, type = "l", col = "blue", xlab = "Month", ylab = "Temperature",
   main = "Temperature chart", ylim=c(-7,35) )
lines(SummarisedTemps$month,SummarisedTemps$max, type = "l", col = "red") 
```
